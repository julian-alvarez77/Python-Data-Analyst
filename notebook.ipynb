**Mengatasi Missing Values pada Orders Dataset**

```python
print('Info Data Orders')
orders_dataset_df.isna().sum()
```

```python
# Pastikan semua kolom waktu dikonversi ke tipe datetime
orders_dataset_df["order_purchase_timestamp"] = pd.to_datetime(orders_dataset_df["order_purchase_timestamp"], errors="coerce")
orders_dataset_df["order_approved_at"] = pd.to_datetime(orders_dataset_df["order_approved_at"], errors="coerce")
orders_dataset_df["order_delivered_carrier_date"] = pd.to_datetime(orders_dataset_df["order_delivered_carrier_date"], errors="coerce")
orders_dataset_df["order_delivered_customer_date"] = pd.to_datetime(orders_dataset_df["order_delivered_customer_date"], errors="coerce")

# Hitung rata-rata waktu
avg_purchase_to_approved = (orders_dataset_df["order_approved_at"] - orders_dataset_df["order_purchase_timestamp"]).mean()
avg_approved_to_carrier = (orders_dataset_df["order_delivered_carrier_date"] - orders_dataset_df["order_approved_at"]).mean()
avg_carrier_to_customer = (orders_dataset_df["order_delivered_customer_date"] - orders_dataset_df["order_delivered_carrier_date"]).mean()

# Isi nilai yang hilang
orders_dataset_df["order_approved_at"].fillna(
    orders_dataset_df["order_purchase_timestamp"] + avg_purchase_to_approved, inplace=True
)
orders_dataset_df["order_delivered_carrier_date"].fillna(
    orders_dataset_df["order_approved_at"] + avg_approved_to_carrier, inplace=True
)
orders_dataset_df["order_delivered_customer_date"].fillna(
    orders_dataset_df["order_delivered_carrier_date"] + avg_carrier_to_customer, inplace=True
)

# Format kembali ke string setelah semua operasi selesai
orders_dataset_df["order_purchase_timestamp"] = orders_dataset_df["order_purchase_timestamp"].dt.strftime("%Y-%m-%d %H:%M:%S")
orders_dataset_df["order_approved_at"] = orders_dataset_df["order_approved_at"].dt.strftime("%Y-%m-%d %H:%M:%S")
orders_dataset_df["order_delivered_carrier_date"] = orders_dataset_df["order_delivered_carrier_date"].dt.strftime("%Y-%m-%d %H:%M:%S")
orders_dataset_df["order_delivered_customer_date"] = orders_dataset_df["order_delivered_customer_date"].dt.strftime("%Y-%m-%d %H:%M:%S")

# Simpan dataset yang telah diperbaiki
orders_dataset_df.to_csv("updated_orders_dataset.csv", index=False)

print("Dataset berhasil diperbaiki dan disimpan.")
print('Info Data Orders')
orders_dataset_df.isna().sum()
```

### Cleaning all data

```python
def analyze_table(df):
    # Print info table
    print(f"\n---Info Tabel---\n")
    df.info()

    # Print missing values table
    print(f"\n---Missing Value---\n")
    print(df.isna().sum())

    # Print duplicated values count
    print(f"\n---Duplikat---\n")
    print(df.duplicated().sum())

    # Print table description
    print(f"\n---Deskripsi Tabel---\n")
    print(df.describe())

analyze_table(all_df)
```

```python
# Menggantikan missing values dengan mean pada kolom numerik
numeric_columns = ['order_estimated_delivery_date', 'payment_sequential', 'payment_installments', 'payment_value', 'review_score',
                    'product_photos_qty', 'price', 'freight_value']
for col in numeric_columns:
    mean_value = all_df[col].mean()
    all_df[col].fillna(mean_value, inplace=True)
```

```python
# Menggantikan missing values dengan mode pada kolom kategori
categorical_columns = ['payment_type', 'product_category_name', 'seller_city']
for col in categorical_columns:
    mode_value = all_df[col].mode()[0]
    all_df[col].fillna(mode_value, inplace=True)
```

```python
# Menggantikan missing values dengan forward fill pada kolom tanggal
date_columns = ['order_approved_at', 'order_delivered_carrier_date']
for col in date_columns:
    all_df[col].fillna(method='ffill', inplace=True)
```

#### 2. Bagaimana performa penjualan dan revenue perusahaan tiap tahun nya? dan berikan Revenue Perbulan nya pada Tahun terakhir Data

```python
# Pastikan kolom 'order_approved_at' berbentuk datetime
all_df['order_approved_at'] = pd.to_datetime(all_df['order_approved_at'])

# Tambahkan kolom tahun dari 'order_approved_at'
all_df['year'] = all_df['order_approved_at'].dt.year

# Hitung pendapatan per tahun
annual_revenue_df = all_df.groupby('year').agg({
    "price": "sum"
}).reset_index()

# Ganti nama kolom agar lebih deskriptif
annual_revenue_df.rename(columns={"price": "revenue"}, inplace=True)

# Tampilkan data pendapatan tahunan
print(annual_revenue_df.head())
```

```python
# Visualisasi menggunakan Matplotlib
plt.figure(figsize=(10, 6))
bars = plt.bar(annual_revenue_df['year'], annual_revenue_df['revenue'], color='skyblue')

# Tambahkan keterangan angka di atas setiap batang dengan format BRL
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Posisi x di tengah batang
        height,                             # Posisi y di atas batang
        f'R${height:,.2f}',                 # Format angka dengan simbol BRL
        ha='center',                        # Horizontal alignment
        va='bottom',                        # Vertical alignment
        fontsize=10                         # Ukuran font
    )

# Atur estetika grafik
plt.title("Pendapatan Per Tahun (Brazil Real)", fontsize=16)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Pendapatan (Revenue)", fontsize=12)
plt.xticks(annual_revenue_df['year'], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Tampilkan grafik
plt.show()
```

Dapat diketahui total Revenue nya adalah
- R$50,415 di tahun 2016
- R$6,447,800 di tahun 2017
- R$7,637,140 di tahun 2018

Kesimpulan nya adalah walaupun penjualan pada tahun 2018 belum sepenuh nya lengkap karena baru sampai Bulan september, tapi ini membuktikan bahwa penjualan tahun 2018 adalah yang terbesar dibanding 2017
